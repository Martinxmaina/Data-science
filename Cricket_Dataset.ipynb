```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.espncricinfo.com/records/highest-career-batting-average-282910"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find all tables on the page
all_tables = soup.find_all("table")

# Print the number of tables found
print(f"Found {len(all_tables)} tables on the page.")

# Iterate through each table and print its HTML
for i, table in enumerate(all_tables):
    print(f"\nTable {i + 1}:")
    print(table)
```

```python
if table:
    # Get the header row
    header_row = table.find("thead").find("tr")
    header_cols = [col.text.strip() for col in header_row.find_all("td")]

    # Get data rows
    data_rows = table.find("tbody").find_all("tr")  # Assuming data is in tbody
    data = []
    for row in data_rows:
        cols = [col.text.strip() for col in row.find_all("td")]
        data.append(cols)

    # Create a pandas DataFrame
    df = pd.DataFrame(data, columns=header_cols)
    print(df)
else:
    print("Table not found with the specified class.")
```
